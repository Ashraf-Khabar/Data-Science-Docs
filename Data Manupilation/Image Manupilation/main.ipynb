## How to display the images : 

```python
image_files = [
    "./nature/1.jpg",
    "./nature/2.jpg",
    "./nature/3.jpg",
]

images = []

for img_path in image_files:
    if os.path.exists(img_path):
        images.append(np.array(mpimg.imread(img_path)))
    else:
        print("File not found:", img_path)
```

```python
plt.figure(figsize=(40, 10))
columns = 5

for i, image in enumerate(images[:3]):
    plt.subplot(len(images[:3]) // columns + 1, columns, i + 1)
    plt.imshow(image)

plt.tight_layout()
plt.show()
```

## Convert all images into tensors : 

```python
images[1].shape
```

```python
reshaped_images = []

for image in images:
    reshaped_image = np.resize(image, (image.shape[0], image.shape[1]))
    pil_image = Image.fromarray(reshaped_image) 
    bw_image = pil_image.convert('L')  
    reshaped_images.append(np.array(bw_image))
```

```python
reshaped_images[1].shape
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(reshaped_images[0], extent=[0, 10, 0, 5])
plt.tight_layout()
plt.show()
```
