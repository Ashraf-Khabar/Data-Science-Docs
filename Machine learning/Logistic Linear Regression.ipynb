```python
import numpy as np
import math
```

```python
# Generate sample data
np.random.seed(42)  # For reproducibility
m = 100  # Number of data points
x1 = np.random.rand(m)
x2 = np.random.rand(m)
Y = np.random.randint(0, 2, m)
```

```python
# Initialize weights and other variables
w1, w2, b = 0, 0, 0
dw1, dw2, db = 0, 0, 0
alpha = 0.01  # Learning rate
J = 0
```

```python
# Number of iterations for gradient descent
num_iterations = 1000
```

```python
for iteration in range(num_iterations):
    J = 0
    dw1, dw2, db = 0, 0, 0

    for i in range(m):
        # Forward pass
        z = w1 * x1[i] + w2 * x2[i] + b
        a = 1 / (1 + np.exp(-z))
        J += (Y[i] * np.log(a) + (1 - Y[i]) * np.log(1 - a))

        # Backward pass
        dz = a - Y[i]
        dw1 += dz * x1[i]
        dw2 += dz * x2[i]
        db += dz

    # Average the cost and gradients over all data points
    J /= m
    dw1 /= m
    dw2 /= m
    db /= m

    # Gradient descent update
    w1 = w1 - alpha * dw1
    w2 = w2 - alpha * dw2
    b = b - alpha * db
```

```python
print("Trained weights:")
print("w1:", w1)
print("w2:", w2)
print("b:", b)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Initialize weights and other variables
w1_values = np.linspace(-2, 2, 100)  # Range of values for w1
w2_values = np.linspace(-2, 2, 100)  # Range of values for w2
J_values = np.zeros((len(w1_values), len(w2_values)))
# Create meshgrid for 3D plotting
W1, W2 = np.meshgrid(w1_values, w2_values)

# Plot the 3D surface
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(W1, W2, J_values, cmap='viridis')

# Set labels and title
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('Cost J')
ax.set_title('Cost Function in 3D')

plt.show()
```
