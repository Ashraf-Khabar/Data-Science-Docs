```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Creation a datase :
* The Linear regression is a methimetical machine learning model for simple predictions when we have a simple dataset for $x = [x^{(1)}, \: x^{(2)}, \cdots, \: x^{(m)}]$.
* Let's explain some notations : 
    * We have a dataset of `vectors`, for example we have : $X = \begin{bmatrix} 2 & 7 \\ 6 & 5 \end{bmatrix}$, so we can say that : $X = [x^{(1)}, x^{(2)}]$, with $x^{(1)} = [x_1^{(1)} = 2, x_2^{(1)} = 7]$ and $x^{(2)} = [x_1^{(2)} = 6, x_2^{(2)} = 5]$ .
    * We can extend this for more complicated databases, with $n$ lignes and $m$ features .
* The loss function is a mean square error : $\mathcal{L}(w, b) = \frac{1}{2m}(y - \hat{y})^2$ with $\hat{y}$ is the predicted value and $y$ is the value in the dataset.

```python
np.random.seed(42)  # For reproducibility
m = 100  # Number of data points
x = np.random.rand(m) * 3  # Generating random x values between 0 and 10
noise = np.random.randn(m)  # Generating random noise
```

```python
# Generate y values with a linear relationship to x and add noise
y = 2 * x + 1 + noise

# Create a scatter plot of the data
plt.scatter(x, y, marker='o')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regression Data Set with Linear Relationship')
plt.grid(True)

# Show the plot
plt.show()
```

# Training the model of regression 

* The $b$ is caled the `bias` of model, and the $w$ is the `weith of input` . 
* The $\alpha$ is called the learning rate, this is used for the model performance .
* $NE$ is the number of itterations for `gradient descent algorithme` for training the model .

```python
w = 0.5
b = 0.5
alpha = 0.01
num_iterations = 5000
```

* We note $f(x_1, x_2, \cdots, x_m)$, we call the `gradient opperator` : $\nabla f(x_1, x_2, \cdots, x_m) = (\frac{\partial f}{\partial x_1}, \frac{df }{dx_2}, \cdots, \frac{df}{dx_m})$ .

```python
for iteration in range(num_iterations):
    J = 0
    dw, db = 0, 0

    for i in range(m):
        # Forward pass
        z = w * x[i] + b
        a = z  # For linear regression, the activation is just the linear combination (z)

        # Cost function (mean squared error)
        J += (y[i] - a) ** 2

        # Backward pass
        dz = a - y[i]
        dw += dz * x[i]
        db += dz

    # Average the cost and gradients over all data points
    J /= (2 * m)  # Divided by 2m for mean squared error
    dw /= m
    db /= m

    # Gradient descent update
    w = w - alpha * dw
    b = b - alpha * db
```

```python
plt.scatter(x, y, marker='o', label='Data')
plt.plot(x, w * x + b, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Regression Data Set with Linear Relationship')
plt.legend()
plt.grid(True)
plt.show()
```
