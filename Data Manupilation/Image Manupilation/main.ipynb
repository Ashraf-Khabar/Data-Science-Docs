## How to display the images : 

```python
image_files = [
    "./nature/1.jpg",
    "./nature/2.jpg",
    "./nature/3.jpg",
]

images = []

for img_path in image_files:
    if os.path.exists(img_path):
        images.append(mpimg.imread(img_path))
    else:
        print("File not found:", img_path)
```

```python
images[0].shape
images[1].shape
images[2].shape
```

```python
plt.figure(figsize=(40, 10))
columns = 5

for i, image in enumerate(images[:3]):
    plt.subplot(len(images[:3]) // columns + 1, columns, i + 1)
    plt.imshow(image)

plt.tight_layout()
plt.show()
```
